# Lab | Random Variable Distributions

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
import numpy as np
import scipy.stats as stats
```

```python
# Refer to the files_for_lab/we_fn_use_c_marketing_customer_value_analysis.csv dataset.

# Instructions
# 1. Get the numerical variables from our dataset.
# 2. Check using a distribution plot if the variables fit the theoretical normal or exponential distribution.
# 3. Check if any of the transformations (log-transform, etc.) we have seen up to this point changes the result.
```

```python
# 1. Get the numerical variables from our dataset.

df = pd.read_csv('we_fn_use_c_marketing_customer_value_analysis.csv')
df.head()

num_df = df.select_dtypes(include = (int, float))
num_df.head()
```

```python
num_df.info()
```

```python
# 2. Check using a distribution plot if the variables fit the theoretical normal or exponential distribution.
for col in num_df.columns:
    fig, ax = plt.subplots()
    sns.histplot(num_df[col], kde=True, ax=ax)
    plt.show()
```

```python
# 3. Check if any of the transformations (log-transform, etc.) we have seen up to this point changes the result.
def replace_inf(i):
    if np.isfinite(i):
        return i
    else:
        return np.NAN   
def log_transformer(df, cols_to_log):
    df_copy = df.copy()
    for col in cols_to_log:
        df_copy[col] = np.log(df_copy[col])
        df_copy[col] = list(map(replace_inf, df_copy[col]))
        df_copy[col] = df_copy[col].fillna(df_copy[col].mean())
    return df_copy

def box_cox_transformer(df, cols_to_transform):
    df_copy = df.copy()
    for col in cols_to_transform:
        df_copy[col] = np.where(df_copy[col] <= 0, 0, df_copy[col])
        df_copy[col].replace(0,df[col].mean(), inplace=True)
        transformed_col, _ = stats.boxcox(df_copy[col])
        df_copy[col] = transformed_col
    return df_copy
```

```python
cols_to_log = ['Customer Lifetime Value','Total Claim Amount']
log_df = log_transformer(num_df, cols_to_log)
```

```python
for col in cols_to_log:
    fig, ax = plt.subplots()
    sns.histplot(log_df[col], kde=True, ax=ax)
    plt.show()
```

```python
log_df.tail()
```

```python
cox_df = box_cox_transformer(num_df, cols_to_log)
```

```python
for col in cols_to_log:
    fig, ax = plt.subplots()
    sns.histplot(cox_df[col], kde=True, ax=ax)
    plt.show()
```
